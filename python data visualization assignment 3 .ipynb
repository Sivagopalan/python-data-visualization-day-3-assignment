```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as snb
from numpy.random import randn, randint, uniform, sample
```

```python
fmri=snb.load_dataset('fmri')
fmri.head()
```

```python
#question : Plot using sns.relplot, line plot, box plot on fmri data set
snb.relplot(x='timepoint',y='signal',data=fmri,hue='region')
```

```python
snb.relplot(x='timepoint',y='signal',data=fmri,hue='region',kind='line')
```
